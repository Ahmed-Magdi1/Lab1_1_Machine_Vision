# **Question_2_LAb1**

**1) Import a colored image from your computer (Lab1_1.jpg).**

```python
# Important libraries
import  cv2 as cv  #Define computer vision library
import numpy as np #Define numpy library
from matplotlib import pyplot as plt #Define matplotlib library
```

```python
colored_image = cv.imread('/content/drive/MyDrive/Lab_1_1.jpg') #Read the image
rgb_image = cv.cvtColor(colored_image, cv.COLOR_BGR2RGB) #Convert the image from BGR to RGB
plt.imshow(rgb_image) #Show the image
```

**2) Find the size of your image.**

```python
# Get the image dimensions (height, width, channels). The number of channels indicates the color depth of the image.
height, width, channels = colored_image.shape

# Print the dimensions of the image
print(f"Width: {width}, Height: {height}, Channels: {channels}")
```

**3) Resize the image (Lab1_1.jpg) with three different resolutions (100 x 100
x 3), (200 x 200 x 3), (400 x 400 x 3).**

```python
# Resize the image to (100 x 100)
image_100x100 = cv.resize(colored_image, (100, 100))
image_200x200 = cv.resize(colored_image, (200, 200))
image_400x400 = cv.resize(colored_image, (400, 400))
```

```python
# Function to display images
def display_image(image, title):
    # Convert from BGR to RGB to display correctly in matplotlib
    rgb_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    plt.imshow(rgb_image)  # Display the image
    plt.title(title)  # Set the title
    plt.axis('off')  # Hide axis
    plt.show()  # Show the plot

# Display the resized images
display_image(image_100x100, '100x100 Resolution (3 Channels)')
display_image(image_200x200, '200x200 Resolution (3 Channels)')
display_image(image_400x400, '400x400 Resolution (3 Channels)')
```

**4) Create a colored image with resolution (320 x 320 x 3) and set all pixels to be black then save it with name (Black.jpg).**

```python
new_colored_image_black = np.ones((320, 320, 3), dtype='uint8')  # Create a new 320x320 image with 3 color channels
plt.imshow(new_colored_image_black) # Display the image
plt.title('320x320 Resolution (3 Channels)')  # The title of the image
plt.axis('off')  # Hide axis
plt.show() # Show the image on the screen
cv.imwrite('/content/Black.jpg', new_colored_image_black) # Save the image with name 'Black.jpg'
```

**5) Create the same picture as (2) but set all pixels to be white and save it as (White.jpg).**

```python
new_colored_image_white = np.ones((320, 320, 3), dtype='uint8') * 255 # Create a new 320x320 image with 3 color channels
plt.imshow(new_colored_image_white) # Display the image
plt.title('320x320 Resolution (3 Channels)')  # The title of the image
plt.axis('off')  # Hide axis
plt.show() # Show the image on the screen
cv.imwrite('/content/White.jpg', new_colored_image_white) # Save the image with name 'Black.jpg'
```

**6) Using the image White.jpg, do the following:**

**- Write the text "BUE" in the center of the image using black font. The font size should be approximately two-thirds of the image height.**

```python
# Calculate font scale as two-thirds of the image height
height, width = new_colored_image_white.shape[:2]
font_scale = (2 / 3) * (height/100)  # 2/3 of image height
font_thickness = 3  # Adjust thickness for clarity
font = cv.FONT_HERSHEY_COMPLEX  # Font type

# Text to write
text = "BUE"

# Calculate text size and position
text_size = cv.getTextSize(text, font, font_scale, font_thickness)[0]
text_x = (width - text_size[0]) // 2  # Centered horizontally
text_y = (height + text_size[1]) // 2  # Centered vertically

# Put the "BUE" text in the center of the image in black font
BUE_text = cv.putText(new_colored_image_white, text, (text_x, text_y), font, font_scale, (0, 0, 0), font_thickness)

# Display the image with the text
plt.imshow(new_colored_image_white)
# plt.title('320x320 Resolution (3 Channels) with "BUE"')
plt.axis('off')  # Hide axis
plt.show()
```

**- Draw a black rectangle around the "BUE" text.**

```python
# Draw a black rectangle around the text "BUE"
# Top-left corner and bottom-right corner coordinates for the rectangle
top_left = (text_x - 10, text_y - text_size[1] - 10)  # Padding of 10 pixels around the text
bottom_right = (text_x + text_size[0] + 10, text_y + 10)

# Draw the rectangle
Rectangle_image = cv.rectangle(BUE_text, top_left, bottom_right, (0, 0, 0), thickness=2)

# Display the image with the text and rectangle
plt.imshow(Rectangle_image)
# plt.title('320x320 Resolution (3 Channels) with "BUE" and Rectangle')
plt.axis('off')  # Hide axis
plt.show()
```

**- Draw black circles near the borders of the image.**

```python
# Draw black circles near the borders of the image
circle_radius = 10  # Radius of the circles
# Coordinates for the circles (corners and center of each edge)
circle_positions = [
    (circle_radius, circle_radius),  # Top-left
    (width - circle_radius, circle_radius),  # Top-right
    (circle_radius, height - circle_radius),  # Bottom-left
    (width - circle_radius, height - circle_radius),  # Bottom-right
    (width // 2, circle_radius),  # Top-center
    (width // 2, height - circle_radius),  # Bottom-center
    (circle_radius, height // 2),  # Left-center
    (width - circle_radius, height // 2)  # Right-center
]

# Draw each circle
for pos in circle_positions:
    circle_image = cv.circle(Rectangle_image, pos, circle_radius, (0, 0, 0), thickness=-1)  # Filled circle


plt.imshow(circle_image)
# plt.title('320x320 Resolution (3 Channels) with "BUE", Rectangle, and Circles')
plt.axis('off')  # Hide axis
plt.show()
```

**-Write your name in blue anywhere on the image.**

```python
# Write your name "AHMED MAGDI" in blue
name = "AHMED MAGDI"
name_font_scale = 0.5  # Scale for the name font size
name_thickness = 2  # Thickness for the name font

# Calculate text size and position for the name
name_size = cv.getTextSize(name, font, name_font_scale, name_thickness)[0]
name_x = 100  # Position from the left edge
name_y = height - 50  # Position from the bottom edge

# Put the name in blue on the image
myname_image = cv.putText(circle_image, name, (name_x, name_y), font, name_font_scale, (0, 0, 255), name_thickness)

# Display the image with the text, rectangle, circles, and name
plt.imshow(myname_image)
# plt.title('320x320 Resolution (3 Channels) with "BUE", Rectangle, Circles, and Name')
plt.axis('off')  # Hide axis
plt.show()
```

**- Draw an ellipse around your name.**

```python
# Draw an ellipse around the name
center_coordinates = (name_x + name_size[0] // 2, name_y - name_size[1] // 2)  # Center of the ellipse
axes_lengths = (name_size[0] // 2 + 10, name_size[1] // 2 + 10)  # Semi-major and semi-minor axes
angle = 0  # Angle of rotation
start_angle = 0  # Starting angle
end_angle = 360  # Ending angle
ellipse_image = cv.ellipse(myname_image, center_coordinates, axes_lengths, angle, start_angle, end_angle, (0, 0, 255), thickness=2)  # Draw the ellipse in blue

# Display the image with the text, rectangle, circles, and name
plt.imshow(ellipse_image)
plt.axis('off')  # Hide axis
plt.show()
```
